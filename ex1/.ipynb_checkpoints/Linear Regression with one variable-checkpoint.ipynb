```python
import numpy as np
X = np.c_[np.ones(df.shape[0]), df['Population']]
y = np.array(df["Profit"])

thetas = np.array([0,0])

print X.shape
print y.shape
print thetas.shape
```

```python
def compute_cost(X, y, theta):
    inner = np.square(((np.dot(X, theta.T)) - y))
    cost = np.sum(inner) / (2 * len(X))
    return cost
```

```python
# Not optimized version
compute_cost(X, y, thetas)
```

```python
def gradient_descent(X, y, theta, alpha, iters):
    J_History = []
    for i in range(iters):
        cost = compute_cost(X, y, theta)
        loss = np.dot(X, theta.T) - y
        gradient = np.dot(X.transpose(), loss)
        theta = theta - ((alpha * gradient) / (len(X)))
        J_History.append(cost)
    return J_History, theta
```

```python
J_History, theta = gradient_descent(X, y, thetas, .01, 1500)
```

```python
new_J_History = {}
for i, v in enumerate(J_History):
    new_J_History[i + 1] = v

cost_df = pd.DataFrame(new_J_History.items(), columns=['Number of Iterations', 'Cost Value'])
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(12,8))  
ax.plot(np.arange(1500), J_History, 'r')
ax.set_xlim([0, 1500])
ax.set_ylim([4, 7])
ax.set_xlabel('Iterations') 
ax.set_ylabel('Cost')  
ax.set_title('Error vs. Training Epoch')  
```
